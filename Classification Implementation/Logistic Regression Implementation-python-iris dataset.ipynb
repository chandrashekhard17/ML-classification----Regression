```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
print(dataset.DESCR)
```

```python
dataset.keys()
```

```python
dataset.feature_names
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
df.head()
```

```python
dataset.keys()
```

```python
print(dataset.target)
```

```python
df['target']=dataset.target
```

```python
df.head()
```

```python
df['target'].unique()
```

We make only binary type of logistic regression classification so we can remove the target value those they containing value 2

```python
df_copy=df[df['target']!=2]
```

```python
df_copy.head()
```

```python
df_copy['target'].unique()
```

we have to decide the dependent and independent variable

```python
x=df_copy.iloc[:,:-1]
y=df_copy.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classification=LogisticRegression(max_iter=200)
```

```python
classification
```

```python
classification.fit(x_train,y_train)
```

```python
y_pred_log=classification.predict(x_test)
```

```python
y_pred_log
```

# Calculating the confusion matrix, accuracy, classification report

```python
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
```

```python
print(confusion_matrix(y_test,y_pred_log))
```

```python
print(accuracy_score(y_test,y_pred_log))
```

```python
print(classification_report(y_test,y_pred_log))
```

# Perform Cross Validation

```python
from sklearn.model_selection import KFold
```

```python
cv=KFold(n_splits=5,shuffle=True)
```

```python
cv1=KFold(n_splits=5,shuffle=False)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores=cross_val_score(classification,x_train,y_train,scoring='accuracy',cv=cv)
```

```python
cross_val_score(classification,x_train,y_train,scoring='accuracy',cv=cv1)
```

```python
final_score=np.mean(scores)
```
