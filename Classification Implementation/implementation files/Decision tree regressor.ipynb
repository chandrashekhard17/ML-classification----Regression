```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# california house pricing dataset
from sklearn.datasets import fetch_california_housing
```

```python
california_df=fetch_california_housing()
```

```python
california_df
```

```python
df=pd.DataFrame(california_df.data,columns=california_df.feature_names)
```

```python
df.head()
```

```python
df['Target']=california_df.target
```

```python
df.head()
```

```python
df.shape
```

```python
#we have to take it sample and then it will be taken and doing gridsearch cv
df=df.sample(frac=0.25)
```

```python
df.head()
```

```python
# independent and dependent features
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor=DecisionTreeRegressor()
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_pred,y_test)
```

```python
score
```

## i got the r2 score value is 51% with selecting the default parameter which is mentioned in the decisio tree regressor itself

```python
parameters={
     'criterion':('squared_error', 'friedman_mse', 'absolute_error', 'poisson'),
     'splitter':('best', 'random'),
     'max_depth':[1,2,3,4,5,6,7,8,10,11,12],
     'max_features':('auto', 'sqrt', 'log2')
}
regressor=DecisionTreeRegressor()
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import GridSearchCV
regressorcv=GridSearchCV(regressor,param_grid=parameters,cv=2,scoring='neg_root_mean_squared_error')
```

```python
regressorcv.fit(x_train,y_train)
```

```python
regressorcv.best_params_
```

```python
regressor=DecisionTreeRegressor(criterion='friedman_mse',max_depth= 6,max_features= 'log2',splitter= 'best')
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_pred,y_test)
```
