# Exploratory Data Analysis

we don't require day month year for eda so we drop it

````python
df_copy=df.drop(['day','month','year'],axis=1)
````

````python
df.copy().head()
````

````python
df_copy['Classes'].value_counts()
````

#encoding of the categories in the class

````python
#the original value of the not fire places replaces 0 otherwise it contains 1
````

````python
df_copy['Classes']=df_copy["Classes"].astype(object)
````

````python
df_copy['Classes'].value_counts()
````

````python
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
````

````python
df_copy.head()
````

````python
df_copy.tail()
````

````python
df_copy['Classes'].value_counts()
````

````python
#plot the density plot for all features 
````

````python
plt.style.use('seaborn')
df_copy.hist(bins=50,figsize=(10,15))
plt.show()
````

````python
#percentage for using pie chart
````

````python
percentage=df_copy['Classes'].value_counts(normalize=True)*100
#plotting pie chart
classlabels=['fire','not fire']
plt.figure(figsize=(7,7))
plt.pie(percentage,labels=classlabels,autopct='%3.2f%%')
plt.title('pie chart of classes')
plt.show()
````

````python
#monthly fire analysis for sidi bellabbes region
df_temp=df.loc[df['region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.ylabel('Number of fires',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title('Fire analysis for sidi bel abbes region',weight='bold')
````

````python
#monthly fire analysis for brjaia regions
df_temp=df.loc[df['region']==0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.ylabel('Number of fires',weight='bold')
plt.xlabel('month',weight='bold')
plt.title('Fire analysis for Brjia regions',weight='bold')
````
